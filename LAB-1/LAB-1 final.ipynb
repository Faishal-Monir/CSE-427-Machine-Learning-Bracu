#**MatPlotLIB**

```python
#Task-1
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')

plt.xlabel('Temperature in Dhaka City')
plt.ylabel('Humidity Based on Temperature Values')
plt.title('Scatter Plot of Temp. vs Humidity')
plt.show()
```

```python
#Task-2
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12, 8))
plt.plot(study_hours,marks,color="k")
plt.title("Study hour vs Marks plot ")
plt.xlabel("Study Hours")
plt.ylabel("Achieved Marks")
plt.show()
```

```python
#Task-3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

plt.bar(subjects, marks, color='steelblue')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title('Subject Wise marks Bar-chart')
plt.grid(axis='y', linestyle='dashdot')
plt.show()
```

```python
#Task-4
colors_data= ['steelblue', 'skyblue', 'teal', 'slategray', 'silver']
plt.barh(subjects, marks, color=colors_data)
plt.ylabel("Subjects")
plt.xlabel("Marks")
plt.title('Marks in Each Subject (Vertical Bar Graph)')
plt.show()
```

```python
#Task-5
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks2 = [89, 90, 45, 78, 99]

colors_data= ['steelblue', 'skyblue', 'teal', 'slategray', 'silver']

plt.figure(figsize=(20, 15))

plt.subplot(2,2,1)
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.xlabel('Temperature in Dhaka City')
plt.ylabel('Humidity of the Temperature Values')
plt.title('Scatter Plot of Temp. vs Humidity')

plt.subplot(2,2,2)
plt.plot(study_hours,marks,color="k")
plt.title("Study hour vs Marks plot ")
plt.xlabel("Study Hours")
plt.ylabel("Achieved Marks")

plt.subplot(2,2,3)
plt.bar(subjects, marks2, color='steelblue')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title('Subject Wise marks Bar-chart')
plt.grid(axis='y', linestyle='dashdot')

plt.subplot(2,2,4)
plt.barh(subjects, marks2, color=colors_data)
plt.ylabel("Subjects")
plt.xlabel("Marks")
plt.title('Marks in Each Subject (Vertical Bar Graph)')

plt.show()

```
