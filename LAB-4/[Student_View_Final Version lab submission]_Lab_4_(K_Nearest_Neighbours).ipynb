# Implementing KNN Classifier from Scratch

```python
class KNN:
    def __init__(self, k):
        self.knn=k
        self.x_train_arr=None
        self.y_train_arr=None

    def euclidean_distance(self, point1, point2):
        eu=np.sqrt(np.sum((np.array(point1) - np.array(point2)) ** 2))
        return eu


    def fit(self, X_train, y_train):
        self.x_train_arr=np.array(X_train)
        self.y_train_arr=np.array(y_train)


    def predict(self, X_test):
        results=[]
        for i in X_test:
            results.append(self.predict_label(i))
        return np.array(results)



    def predict_label(self, x):
        distance=[]
        for i in self.x_train_arr:
            distance.append(self.euclidean_distance(x,i ))

        sorting=np.argsort(distance)
        neighbors=sorting[:self.knn]
        # print(neighbors)

        ytrain_labels=[]
        for j in neighbors:
            ytrain_labels.append(self.y_train_arr[j])

        #database_for majority voting

        label_counter={}

        for k in ytrain_labels:
            if k in label_counter:
                label_counter[k]+=1
            else:
                label_counter[k]=1

        # print(label_counter)

        #voting
        winner=None
        counter=0

        for label,votes in label_counter.items():
            if votes>counter:
                counter=votes
                winner=label

        return winner
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=41)

# Initialize KNN classifier with the value of n
knn = KNN(3)

# Fit the model
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

counter=0

for i in range(len(y_test)):
    if y_test[i]==y_pred[i]:
        counter+=1
accuracy=counter/len(y_test)
print(f"Accuracy of KNN from scratch: {accuracy * 100:.2f}%")
```
